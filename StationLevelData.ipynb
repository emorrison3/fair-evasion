```r
library(dplyr)
library(readr)
stations_arrests <- read_csv("clean arrests.csv") #read in arrest data 
station_local_demo_a <- read_csv("local_station_demo.csv") #read in neighborhood demographic data
station_local_demo_a <- subset(station_local_demo, select=c(AI, API, Black, Hispanic, Unknown, White)) #drop 'Total' column
arrests_matrix <- data.matrix(stations_arrests, rownames.force = NA)
demo_matrix_a <- data.matrix(station_local_demo_a, rownames.force = NA) #turn dfs into matrices to perform operations
```

```r


b_test <- function(x, n, p){   ##perform binomial test on the number of summons given to a specific group (x) out of the total number of summons 
    ##for that sation (n) comparing it to that group's representation in the local area (p)
  
    binom.test(x, n, p)
    } 

print(mapply(b_test, arrests_matrix[,1:6], arrests_matrix[,7], demo_matrix_a[,1:6])) 
##allows rowwise application of b_test function
```
